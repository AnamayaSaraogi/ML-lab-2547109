#### For dataset city_day.csv

```python
# Create a copy so the original dataset remains unchanged
city_day_imputed = city_day.copy()

# Drop Xylene
city_day_imputed.drop('Xylene', axis=1, inplace=True)

# Drop rows with missing target values
city_day_imputed.dropna(subset=['AQI', 'AQI_Bucket'],inplace=True)

# Median imputation
num_cols = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene']

for col in num_cols:
    city_day_imputed[col] = city_day_imputed[col].fillna(
        city_day_imputed[col].median()
    )

# Verify
print(city_day_imputed.isnull().sum())
```

**Missing Value Treatment Strategy**

**Xylene**
The Xylene column contains more than 60% missing values. Since a majority of the data is unavailable, imputing such a large number of values would introduce significant uncertainty and may not represent the true distribution of the data. Therefore, this column was dropped.

**AQI and AQI_Bucket**
AQI and AQI_Bucket are the target variables of the dataset. Imputing missing values in the target variable can lead to inaccurate and misleading results during model training. Since only a relatively small portion of records had missing values in these columns, those records were removed.

**PM2.5, PM10, NO, NO2, NOx, NH3, CO, SO2, O3, Benzene and Toluene**
These columns contain numerical pollutant measurements. Instead of removing rows, the missing values were replaced with the median of the respective column. The median was chosen because environmental data often contains extreme values and outliers, and the median is less affected by such values than the mean. This helps preserve the dataset while maintaining realistic values.

After applying the above treatment:
- The Xylene column was removed.
- Rows with missing AQI and AQI_Bucket values were dropped.
- Missing values in the remaining numerical columns were imputed using the median.

A final null-value check confirmed that no missing values remain in the cleaned dataset.

### Task 5 Something is making the average AQI look worse than it is


A data quality report flags that a few AQI readings in the dataset are implausibly high — values that no monitoring station should realistically record. If left in, these values will distort every statistic and model built on this data. The team lead asks you to “handle the extreme values properly” but does not tell you how.
Identify whether extreme values exist, quantify them, decide on an appropriate treatment method, apply it, and demonstrate that the data is now cleaner. Justify every decision you make.

The count of values affected and the treatment applied

```python
# detecting outliers using IQR method

Q1 = city_day_imputed['AQI'].quantile(0.25)
Q3 = city_day_imputed['AQI'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = city_day_imputed[
    (city_day_imputed['AQI'] < lower_bound) |
    (city_day_imputed['AQI'] > upper_bound)
]

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Number of Outliers:", len(outliers))
```

```python
# treating outliers using Winsorization

city_day_clean = city_day_imputed.copy()

city_day_clean['AQI'] = city_day_clean['AQI'].clip(
    lower=lower_bound,
    upper=upper_bound
)
```

```python
# Visual comparison of AQI before and after outlier treatment

import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

sns.boxplot(x=city_day_imputed['AQI'], ax=axes[0], color='skyblue')
axes[0].set_title("AQI Before Outlier Treatment")
axes[0].set_xlabel("AQI")
axes[0].set_ylabel("AQI")

sns.boxplot(x=city_day_clean['AQI'], ax=axes[1], color='salmon')
axes[1].set_title("AQI After Outlier Treatment")
axes[1].set_xlabel("AQI")

plt.tight_layout()
plt.show()
```

```python
affected = (
    city_day_imputed['AQI'] > upper_bound
).sum()

print("Values capped:", affected)
```

**Outlier Detection Method**

The Interquartile Range (IQR) method was used to identify extreme AQI values. This method was chosen because the AQI distribution is highly right-skewed and contains several unusually large values. Unlike methods based on the mean and standard deviation, the IQR method is less affected by extreme observations and is therefore more suitable for skewed environmental data.

The lower and upper bounds were calculated as:

- Lower Bound = Q1 − 1.5 × IQR
- Upper Bound = Q3 + 1.5 × IQR

Any AQI value outside these bounds was considered an outlier.

**Treatment Applied**

Rather than removing outliers, Winsorization (capping) was applied. Values above the upper bound were replaced with the upper-bound value.

This approach was chosen because extreme AQI readings may still correspond to genuine pollution events. Deleting them would result in information loss, whereas capping reduces their influence on summary statistics while preserving all observations.

**Results**

- Number of AQI values identified as outliers: 1358
- Treatment applied: Capping using the IQR upper bound

**Evidence of Improvement**

The box plots before and after treatment show a substantial reduction in the influence of extreme AQI values. The distribution remains representative of the data while preventing a small number of unusually high observations from disproportionately affecting statistical analyses and machine learning models.

The analysis shows a clear seasonal pattern in air quality throughout the year.
```python
# calculate average AQI by state
state_aqi = (
    city_day_clean
    .groupby('State', as_index=False)['AQI']
    .mean()
)

state_aqi.head()
```

```python
# calculate average crop production by state
state_crop = (
    crop_prod_imputed
    .groupby('State_Name', as_index=False)[['Area', 'Production']]
    .mean()
)

state_crop.rename(columns={'State_Name':'State'}, inplace=True)

state_crop.head()
```

```python
# merge the two datasets on State column
merged_df = pd.merge(
    state_crop,
    state_aqi,
    on='State',
    how='inner'
)

print("Merged Shape:", merged_df.shape)
merged_df.head()
```

```python
# correlation matrix of numerical variables in the merged dataset
import seaborn as sns
import matplotlib.pyplot as plt

corr = merged_df.select_dtypes(include='number').corr()

plt.figure(figsize=(6,5))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    center=0
)

plt.title("Correlation Matrix of Numerical Variables")

plt.show()
```

**Transformation Before Merge**

The air quality dataset was originally recorded at the city-day level, while the crop production dataset was recorded at the state-year level. A direct merge on State and Year was not possible because the datasets did not contain sufficient overlapping state-year records.

To make the datasets compatible, both datasets were aggregated to the state level. The average AQI was calculated for each state from the air quality dataset, while the average cultivated area and average crop production were calculated for each state from the crop dataset. This produced one record per state in both datasets, allowing them to be merged using the State column.

**Relationships Identified**

1. AQI and Crop Production (Correlation = -0.22)

The correlation between AQI and crop production is **-0.22**, indicating a weak negative relationship. States with higher average air pollution tend to have slightly lower crop production. One possible explanation is that poor air quality can reduce the amount of sunlight reaching crops and may negatively affect plant growth and photosynthesis. However, the relationship is weak, suggesting that crop production is also influenced by many other factors such as rainfall, irrigation, soil fertility, crop type, and farming practices.

2. Area and AQI (Correlation = 0.25)

The correlation between cultivated area and AQI is **0.25**, indicating a weak positive relationship. States with larger agricultural areas tend to have slightly higher AQI values. This may be because states with extensive agricultural land often experience greater agricultural activities, including machinery use and seasonal crop residue burning, which can contribute to higher pollution levels. However, the relationship is relatively weak and does not imply that larger cultivated areas directly cause poor air quality.
