# EDA stands for Exploratory Data Analysis.         :

# The main goal is to answer questions like:

What does the data look like?

Are there missing values?

Are there outliers?

Which features are important?

```python
titanic_drop = titanic.dropna(subset=['age'])

print("Original Shape:", titanic.shape)
print("After Removing Missing Ages:", titanic_drop.shape)
```

Understanding Each Operation
copy()

Creates a duplicate dataset.

Without this, the original dataset changes permanently.

mean()

Calculates average.

---



Mean formula:

xˉ=∑x/n
	​


Where:

∑x = sum of all values

n = number of values




Example:

Ages = 20, 30, 40

Mean: (20+30+40)/3=30

fillna(mean_age)

Replaces missing values using the calculated average.

Example:

Age Before

If mean = 29:

Why Mean Filling Helps

Advantages:

Keeps all rows
Simple and fast

Disadvantage:

Can reduce variability
May introduce bias

**Method 3: Fill Missing Categorical Values Using Mode**
mode_town = titanic['embark_town'].mode()[0]

titanic['embark_town'] = titanic['embark_town'].fillna(mode_town)

**Understanding Mode**

Mode = most frequent value.

Example:

City
Chennai
Bangalore
Chennai

Mode = Chennai

```python
titanic_mean = titanic.copy()

mean_age = titanic_mean['age'].mean()

titanic_mean['age'] = titanic_mean['age'].fillna(mean_age)

print("Missing Values After Filling:")
print(titanic_mean['age'].isnull().sum())
```

Outlier Detection
What is an Outlier?

An outlier is an unusually different value.

Example:

Ages
20
25
23
200

200 is suspicious.

```python
Q1 = titanic['age'].quantile(0.25)
Q3 = titanic['age'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = titanic[(titanic['age'] < lower_bound) |
                   (titanic['age'] > upper_bound)]

print(outliers[['age']].head())
```

# **Gender Wise Survival**

Understanding Countplot

Shows frequency of categories.

Here:

1 → Survived
Interpretation

```python
sns.countplot(x='survived', data=titanic)

plt.title("Survival Count")

plt.show()
```

Correlation Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation = titanic.corr(numeric_only=True)

plt.figure(figsize=(10,8))

sns.heatmap(correlation,
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```
